```
#########################################################################################
#                                Calculator Program                                     #
#########################################################################################

#import log function from math library
from math import log

#squaring function
#input: number entered by user
#output: square of the number
def squaring(int_A):
  return int_A ** 2

#natural log function
#input: number entered by user
#output: natural log of the number
def natural_log(int_A):
  return log(int_A)

#absolute value function
#input: number entered by user
#output: absolute value of the number
def absolute_value(int_A):
  return abs(int_A)

#addition function
#input: 2 numbers entered by user
#output: sum of the numbers
def addition(int_A, int_B):
  return int_A + int_B

#substraction function
#input: 2 numbers entered by user
#output: substraction of the numbers
def substraction(int_A, int_B):
  return int_A - int_B

#multiplication function
#input: 2 numbers entered by user
#output: multiplication of the numbers
def multiplication(int_A, int_B):
  return int_A * int_B

#division function
#input: 2 numbers entered by user
#output: division of the numbers
def division(int_A, int_B):
  return int_A / int_B

#exponentiation function
#input: 2 numbers entered by user
#output: exponentiation of the numbers
def exponentiation(int_A, int_B):
  return int_A ** int_B

#modulus function
#input: 2 numbers entered by user
#output: modulus of the numbers
def modulus(int_A, int_B):
  return int_A % int_B

#check_if_number function
#input: number entered by user
#output: 0 or 1 based if the input is a valid number
def check_if_number(number_in):
    try:
      val = float(number_in)
      return 1
    except ValueError:
      return 0

#check_operator function
#input: operator entered by user
#output: 0 or 1 based if the input is a valid operator
def check_operator(operator):
  if operator == 'abs' or operator == '+' or operator == '/' or operator == '^' or operator == '%':
    return 1
  elif operator == '*' or operator == 'log' or operator == 'sqr' or operator == '-':
    return 1
  else:
    return 0

#help function
#input: no input
#output: displays supported operations message to screen
def help():
  print("\nSupported Operations ['+', '-', '*', '/', '^', 'abs', 'log', 'sqr']")

#good_bye function
#input: no input
#output: displays good bye message to screen
def good_bye():
  print("\nThank you for using this program and have a great day.")
  print("Good Bye.")

#intro function
#input: no input
#output: displays welcome message to screen
def intro():
  print("           ************************************************************")
  print("           *****************Welcome to Simple Calulator****************")
  print("           ************************************************************")

#########################################################################################
#                                    Start of Program                                   #
#########################################################################################

#variable declaration and initialition. variable ends execution on program main loop
stop = 0
intro()

#main while loop of the program. when variable stop = 1 program ends
while stop != 1:

###################################################################################
# This segment of code requests a number from the user or 'x' to exit the program #
# If input is 'x' the program ends. Else, block of code will check if the input   #
# is a valid number. When input is valid, program will continue                   #                                                        #
###################################################################################
  print("\nPlease enter the first number to compute")
  int_A = input("or the letter 'x' to exit the program: ")
  if int_A == 'x':
    stop = 1
    good_bye()
    continue
  else:
    num_chk = check_if_number(int_A)
    while num_chk != 1:
      print("\nThe entered data is not a number.")
      int_A = input("Please enter a valid number (option to end program will be provided after this step): ")
      num_chk = check_if_number(int_A)
###################################################################################

###################################################################################
# This segment of code prints the list of available operators, requests the       #
# operator from the user or 'x' to exit the program. If input is 'x' the program  #
# ends. Else, block of code will check if the input is a valid operator. When     #
# input is valid, program will continue                                           #
###################################################################################
  help()
  print("Please enter the operation to be executed")
  operator = input("or the letter 'x' to exit the program: ")
  if operator == 'x':
    stop = 1
    good_bye()
    continue
  else:
    oper_chk = check_operator(operator)
    while oper_chk != 1:
      help()
      print("\nThe entered data is not an operator.")
      operator = input("Please enter a valid operator from the list (option to end program will be provided after this step): ")
      oper_chk = check_operator(operator)
###################################################################################

###################################################################################
# This segment of code is a conditional statement which checks for specific       #
# operators which can be completed and executed with only one number. If user     #
# requests absolute value, log, or squaring, the block of code executes the       #
# operation, and skips the rest of the program                                    #
###################################################################################
    if operator == 'abs':
      print(operator + " of " + int_A + " is: %.2f\n" % float(absolute_value(float(int_A))))
      continue
    elif (operator == 'log'):
      print(operator + " of " + int_A + " is: %.2f\n" % float(natural_log(float(int_A))))
      continue
    elif operator == 'sqr':
      print(operator + " of " + int_A + " is: %.2f\n" % float(squaring(float(int_A))))
      continue
###################################################################################

###################################################################################
# This segment of code requests a number from the user or 'x' to exit the program #
# If input is 'x' the program ends. Else, block of code will check if the input   #
# is a valid number. When input is valid, program will continue                   #
###################################################################################
  print("\nPlease enter the second number to compute")
  int_B = input("or the letter 'x' to exit the program: ")
  if int_B == 'x':
    stop = 1
    good_bye()
    continue
  else:
    num_chk = check_if_number(int_B)
    while num_chk != 1:
      print("\nThe entered data is not a number.")
      int_B = input("Please enter a valid number (option to end program will be provided after this step): ")
      num_chk = check_if_number(int_B)
###################################################################################

###################################################################################
# This segment of code is a conditional statement which checks for all other      #
# operators. The block of code will execute the operation entered by the user     #
# based on the input numbers and print result to the screen                       #
###################################################################################
  if operator == '+':
    print("\n" + int_A + " " + operator + " " + int_B + " is: %.2f\n" % float(addition(float(int_A), float(int_B))))
  elif operator == '-':
    print("\n" + int_A + " " + operator + " " + int_B + " is: %.2f\n" % float(substraction(float(int_A), float(int_B))))
  elif operator == '*':
    print("\n" + int_A + " " + operator + " " + int_B + " is: %.2f\n" % float(multiplication(float(int_A), float(int_B))))
  elif operator == '/':
    print("\n" + int_A + " " + operator + " " + int_B + " is: %.2f\n" % float(division(float(int_A), float(int_B))))
  elif operator == '^':
    print("\n" + int_A + " " + operator + " " + int_B + " is: %.2f\n" % float(exponentiation(float(int_A), float(int_B))))
  elif operator == '%':
    print("\n" + int_A + " " + operator + " " + int_B + " is: %.2f\n" % float(modulus(float(int_A), float(int_B))))
###################################################################################
```
